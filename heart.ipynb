```python
def classify_features(df):
    categorical_features = []
    non_categorical_features = []
    discrete_features = []
    continuous_features = []

    for column in df.columns:
        if df[column].dtype == 'object':
            if df[column].nunique() < 10:
                categorical_features.append(column)
            else:
                non_categorical_features.append(column)
        elif df[column].dtype in ['int64']:
            if df[column].nunique() < 10:
                discrete_features.append(column)
            else:
                continuous_features.append(column)

    return categorical_features, non_categorical_features, discrete_features, continuous_features
```

```python
categorical, non_categorical, discrete, continuous = classify_features(df)
```

```python
print("Categorical Features:", categorical)
print("Non-Categorical Features:", non_categorical)
print("Discrete Features:", discrete)
print("Continuous Features:", continuous)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
df_new =df.copy()
```

```python
X = df_new.drop('DEATH_EVENT', axis=1)
y = df_new['DEATH_EVENT']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42)
```

```python
df.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE()
X_re_train, y_re_train = smote.fit_resample(X_train, y_train)
```

```python
log_model = LogisticRegression()
log_model.fit(X_re_train, y_re_train)
```

```python
log_pred = log_model.predict(X_test)
log_accuracy_m = accuracy_score(y_test, log_pred)
print("Logistic Regression Accuracy:", log_accuracy_m)
```

```python
log_train = log_model.predict(X_train)
log_accuracy = accuracy_score(y_train, log_train)
print("Logistic Regression Training  Accuracy:", log_accuracy)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X_re_train, y_re_train)
```

```python
rf_pred = rf_model.predict(X_test)
rf_accuracy_m = accuracy_score(y_test, rf_pred)
print("Random Forest Accuracy:", rf_accuracy_m)
```

```python
rf_train=rf_model.predict(X_train)
rf_accuracy=accuracy_score(y_train,rf_train)
print("Random Forest training Accuracy:", rf_accuracy)
```

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_re_train, y_re_train)
```

```python
dt_pred = dt_model.predict(X_test)
dt_accuracy_m = accuracy_score(y_test, dt_pred)
print("Decision Tree Accuracy:", dt_accuracy_m)
```

```python
dt_train = dt_model.predict(X_train)
dt_accuracy = accuracy_score(y_train, dt_train)
print("Dt Training Accuracy:",dt_accuracy)
```

```python
svm_model = SVC()
svm_model.fit(X_re_train, y_re_train)
```

```python
svm_pred = svm_model.predict(X_test)
svm_accuracy = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_accuracy)
```

```python
svm_train = svm_model.predict(X_train)
train_accuracy = accuracy_score(y_train, svm_train)
print("Training Accuracy:", train_accuracy)
```

```python
svm_model = SVC(kernel='linear')
svm_model.fit(X_re_train, y_re_train)
```

```python
svm_pred = svm_model.predict(X_test)
svm_accuracy_m = accuracy_score(y_test, svm_pred)
print("SVM Accuracy:", svm_accuracy_m)
```

```python
svm_train = svm_model.predict(X_train)
train_accuracy = accuracy_score(y_train, svm_train)
print("Training Accuracy:", train_accuracy)
```
